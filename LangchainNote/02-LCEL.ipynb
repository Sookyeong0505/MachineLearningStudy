# LangChain Expression Languages(LCEL)

**기본 예시: 프롬프트 + 모델 + 출력 파서**

가장 기본적이고 일반적인 사용 사례는 prompt 템플릿과 모델을 함께 연결하는 것이다.

예시) 각 나라별 대통령을 물어보는 Chain을 생성


## 프롬프트 템플릿(PromptTemplate)
사용자의 입력 변수를 사용하여 완전한 프롬프트 문자열을 만드는 템플릿.

**사용법**
- `template`: 템플릿 문자열. 이 문자열 내에서 중괄호 {}는 변수를 나타낸다. {} 안에 변수가 들어간다.
- `input_variables`: 중괄호 안에 들어갈 변수의 이름을 리스트로 정의.

```python
from langchain_teddynote.messages import stream_response  # 스트리밍 출력
from langchain_core.prompts import PromptTemplate

# template 정의
template = "{country}의 대통령은 누구인가요?"

# from_template 메소드를 이용하여 PromptTemplate 객체 생성
prompt_template = PromptTemplate.from_template(template)
prompt_template
```

```python
# prompt 생성
prompt = prompt_template.format(country="대한민국")
prompt
```

```python
# prompt 생성
prompt = prompt_template.format(country="미국")
prompt
```

## Chain
LCEL을 사용하여 다양한 구성 요소를 단일 체인으로 결합할 수 있다.

체인은 유닉스 파이프 연산자와 유사한 `|` 기호를 사용해서 서로 다른 구성 요소를 연결해 **한 구성 요소의 출력을 다음 구성 요소의 입력으로 전달**한다.

`chain = prompt | model | output_parser`

체인의 흐름: 사용자 입력 -> 프롬프트 -> 모델 -> 출력 파서

`chain.invoke(input)` 메서드를 사용하여 체인을 실행하고, input 딕셔너리를 전달하여 결과를 얻을 수 있다.

```python
from langchain_openai import ChatOpenAI

# prompt 를 PromptTemplate 객체로 생성.
prompt = PromptTemplate.from_template("{topic}에 대해 설명해주세요.")

model = ChatOpenAI(
    model="gpt-3.5-turbo",
    max_tokens=2048,
    temperature=0.1,
)

chain = prompt | model
```

```python
# input 딕셔너리에 주제 설정
input = {"topic": "LangChain"}

# prompt 객체와 model 객체를 파이프(|) 연산자로 연결하고 invoke 메서드를 사용하여 input을 전달한다.
# 이를 통해 AI 모델이 생성한 메시지를 반환합니다.
chain.invoke(input)
```

```python
# 스트리밍 출력
answer = chain.stream(input)
stream_response(answer)
```

## 출력 파서(Output Parser)


```python
from langchain_core.output_parsers import StrOutputParser

output_parser = StrOutputParser()

# 프롬프트, 모델, 출력 파서를 연결하여 처리 체인을 구성합니다.
chain = prompt | model | output_parser

# chain 객체의 invoke 메서드를 사용하여 input을 전달합니다.
input = {"topic": "LangChain"}
chain.invoke(input)
```

```python
# 스트리밍 출력
answer = chain.stream(input)
stream_response(answer)
```

## 템플릿 변경해서 적용하기
프롬프트의 내용을 변경하거나 model의 파라미터를 변경하여 다양한 체인을 구성할 수 있다.

```python
template = """
당신은 영어를 가르치는 10년차 영어 선생님입니다. 상황에 [FORMAT]에 영어 회화를 작성해 주세요.

상황:
{question}

FORMAT:
- 영어 회화:
- 한글 해석:
"""

# 프롬프트 템플릿을 이용하여 새로운 프롬프트 생성.
prompt = PromptTemplate.from_template(template)

# ChatOpenAI 모델을 초기화한다. 새로운 모델 넣어보기.
model = ChatOpenAI(model_name="gpt-4-turbo")

# 문자열 출력 파서를 초기화.
output_parser = StrOutputParser()

# 체인 구성
chain = prompt | model | output_parser
```

```python
# chain을 실행하여 스트림 답변 얻기
answer = chain.stream({"question": "저는 식당에 가서 음식을 주문하고 싶어요"})
# 스트리밍 출력
stream_response(answer)
```

```python
answer = chain.stream({"question": "영국에서 피쉬앤칩스 주문하며 가게 주인과 스몰토킹하는 상황의 영어회화를 만들어주세요."})
stream_response(answer)
```

## Runnable 프로토콜
사용자의 입력에 따라 체인을 정의하고 실행할 수 있는 기능을 제공하는 프로토콜이다.

표준 인터페이스로서, 체인을 정의하고 실행하는 데 필요한 메서드를 제공한다.

- `invoke(input: dict) -> Any`: 체인을 실행하고 결과를 반환하는 메서드.
- `stream(input: dict) -> Any`: 체인을 실행하고 스트리밍 결과를 반환하는 메서드.
- `batch(inputs: List[dict]) -> List[Any]`: 리스트로 여러 입력을 한 번에 받아 처리하는 메서드.

비동기 메서드
- `ainvoke(input: dict) -> Any`: 체인을 비동기적으로 실행하고 결과를 반환하는 메서드.
- `astream(input: dict) -> Any`: 체인을 비동기적으로 실행하고 스트리밍 결과를 반환하는 메서드.
- `abatch(inputs: List[dict]) -> List[Any]`: 여러 입력을 비동기적으로 처리하는 메서드.
- `astream_log(input: dict) -> Any`: 체인을 비동기적으로 실행하고 중간 단계의 스트리밍 로그를 반환하는 메서드.

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

# ChatOpenAI 모델을 인스턴스화
model = ChatOpenAI()

# 프롬프트 템플릿 생성
prompt = PromptTemplate.from_template("{topic}에 대한 농담을 하나 말해줘.")

# 출력 파서 초기화
output_parser = StrOutputParser()

# 체인 생성
chain = prompt | model | output_parser
```

```python
# invoke() 메서드로 체인 실행
chain.invoke({"topic": "컴퓨터 메모리"})
```

```python
# chain.batch() 메서드로 여러 입력을 한 번에 처리
chain.batch([{"topic": "개발자"}, {"topic": "인공지능"}, {"topic": "사랑"}])
```

```python
# chain.batch() 메서드의 config 매개변수는 체인을 실행할 때 다양한 설정을 지정할 수 있는 옵션을 제공한다.
# 설정 가능한 옵션
# - `max_concurrency`: 동시에 처리할 수 있는 작업 수
# - `max_tokens`: 모델이 생성할 최대 토큰 수
# - `temperature`: 모델의 창의성 정도를 조절하는 매개변수
# - `top_p`: 샘플링에서 사용할 누적 확률 질량을 설정
# - `stop`: 모델이 출력을 중단할 조건을 설정
# - `logprobs`: 모델이 각 토큰에 대해 계산한 로그 확률
# - `echo`: 모델이 입력을 그대로 출력에 포함할지 여부를 설정
# - `presence_penalty`: 모델이 새로운 토픽을 생성하도록 유도하는 정도

chain.batch(
    [
        {"topic": "한국 경제"},
        {"topic": "호주"},
        {"topic": "우정"},
        {"topic": "세월"},
        {"topic": "기상청"},
    ],
    config={"max_concurrency": 3}, # 동시에 처리할 수 있는 작업 수를 3개로 설정
)
```

```python
# chain.astream() 메서드로 비동기적으로 체인을 실행하고 스트리밍 결과를 반환

# 비동기 for 루프를 사용해 스트림에서 응답을 순차적으로 받는다.
async for token in chain.astream({"topic": "파티"}):
    # 응답을 출력한다. 줄바꿈 없이 바로 출력하고 버퍼를 비운다.
    print(token, end="", flush=True)
```

```python
# chain.ainvoke() 메서드로 비동기적으로 체인을 실행하고 결과를 반환
result = chain.ainvoke({"topic": "주식"})
await result
```

```python
# chain.abatch() 메서드로 여러 입력을 비동기적으로 일괄 처리
# await 키워드 : 비동기 함수의 실행이 완료될 때까지 기다린다.
my_batch_process = chain.abatch(
    [{"topic": "YouTube"}, {"topic": "사계절"}, {"topic": "에어컨"}]
)
await my_batch_process
```

```python
my_batch_process = chain.abatch(
    [{"topic": "왕좌의 게임"}, {"topic": "일본"}, {"topic": "계급"}]
)
```
